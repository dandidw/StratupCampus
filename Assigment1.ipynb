```python
#kenapa lebih memilih display dari pada print https://www.geeksforgeeks.org/display-the-pandas-dataframe-in-table-style/
print(df)
display(df)
```

```python
#merubah tinggi badan dari Centimeter (cm) ke Meter (m)
tinggi_m = df['tinggi_badan'] / 100
display(tinggi_m)
```

```python
#Mencari BMI (Body Mass Index) dengan membagi berat badan dengan pangkat 2 tinggi badan dalam meter (m)
df['BMI'] = df['berat_badan'] / (tinggi_m ** 2)
display(df)
```

```python
#Import Numpy digunakan untuk mengevaluasi kondisi yang terjadi lalu mengisi kolom baru dalam DataFrame berdasarkan kondisi tersebut.
#https://numpy.org/doc/stable/reference/generated/numpy.select.html
import numpy as np
hitung = [
    (df['BMI'] < 18.5),
    (df['BMI'] >= 18.5) & (df['BMI'] <= 24.9),
    (df['BMI'] >= 25.0) & (df['BMI'] <= 29.9),
    (df['BMI'] >= 30.0) & (df['BMI'] <= 34.9)
]
# Membuat variable kategori
kategori = ['Underweight', 'Normal', 'Overweight', 'Obesity']

# Membuat Kolom Status BMI di kolom dengan menerapkan np.select berdasarkan kondisi yang telah di tentukan dalam variabel hitung dan kategori yang telah di tentukan dalam variabel kategori
df['Status BMI'] = np.select(hitung, kategori)
display(df)
```

```python
# Membuat Sorting Dari Tinggi Badan yang tertinggi
df_sorted_tinggi = df.sort_values(by='tinggi_badan', ascending=False)
display(df_sorted_tinggi)
```

```python
# Membuat Sorting dari 'Nama' descending order (Z - A)
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html
df_urutkan_nama = df.sort_values(by='nama', ascending=False)
display(df_urutkan_nama)
```

```python
# Mencari Status BMI yang tidak normal
df_urutkan_status = df[df['Status BMI'] != 'Normal']
display(df_urutkan_status)
```

```python
# Memfilter tinggi badan yang lebih dari 169 cm dan berat badan yang lebih dari 60 kg
df_filter_tinggi_berat = df[(df['tinggi_badan'] > 169) & (df['berat_badan'] > 60)]
display(df_filter_tinggi_berat)
```

```python
# Mencari nama yang mengandung huruf y
# https://www.geeksforgeeks.org/python-pandas-series-str-contains/
df_nama_y = df[df['nama'].str.contains('y')]
display(df_nama_y)
```

```python
# Mencari Berat Badan yang kurang dari 60 kg
df_kurangdari_60 = df[df['berat_badan'] < 60]
display(df_kurangdari_60)
```

```python
# Menambah kolom 'No.'
df.index = np.arange(1, len(df) + 1)
df.index.name = 'No.'
display(df)
```

```python
# Mencari max, min, average
status_tinggi= df['tinggi_badan'].describe()
status_berat = df['berat_badan'].describe()
display(status_tinggi, status_berat)
```

```python
#Mencari Rata-rata Tinggi Badan
rata_tinggi = df['tinggi_badan'].mean()
rata_tinggi_rounded = round(rata_tinggi, 2)
display('Rata-Rata Tinggi Badan: {:.2f}'.format(rata_tinggi_rounded))
```

```python
#Mencari Rata-Rata Berat Badan
rata_berat = df['berat_badan'].mean()
display('Rata-Rata Berat Badan : {:.2f} '.format(rata_berat))
```

```python
#Mencari Jumlah mentee
num_mente = len(df)
display(num_mente)
```

```python
#Mencari 5 nilai tertinggi dari hasil perhitungan bmi
top_bmi = df.nlargest(5, 'BMI')
display(top_bmi)
```
